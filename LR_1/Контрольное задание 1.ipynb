#####  3. Каковы медиана и стандартное отклонение платежей (Fare)?


##### 4. Правда ли, что люди моложе 30 лет выживали чаще, чем люди старше 60 лет? Каковы доли выживших в обеих группах?

Напишите код и правильный вариант ответа

1) 22.7% среди молодых и 40.6% среди старых

2) 40.6% среди молодых и 22.7% среди старых

3) 35.3% среди молодых и 27.4% среди старых

4) 27.4% среди молодых и 35.3% среди старых

```python
# ваш код здесь
print("survivors less than 30: {:.1%}".format(
    df[df['Age'] <= 30]['Survived'].mean()))
print("survivors more than 60: {:.1%}".format(
    df[df['Age'] > 60]['Survived'].mean()))

print("\nthe correct answer is number 2")
```

##### 6. Сравните графически распределение стоимости билетов и возраста у спасенных и у погибших. Средний возраст погибших выше, верно?

Напишите код и правильный вариант ответа

1) Да

2) Нет


```python
# ваш код здесь
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(df[df['Survived'] == 1]['Fare'], bins=20)
plt.title('Survived')
plt.xlabel('Fare')
plt.ylabel('Freq')

plt.subplot(1, 2, 2)
plt.hist(df[df['Survived'] == 0]['Fare'], bins=20)
plt.title('Died')
plt.xlabel('Fare')
plt.ylabel('Freq')

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(df[df['Survived'] == 1]['Age'], bins=20)
plt.title('Survived')
plt.xlabel('Age')
plt.ylabel('Freq')

plt.subplot(1, 2, 2)
plt.hist(df[df['Survived'] == 0]['Age'], bins=20)
plt.title('Died')
plt.xlabel('Age')
plt.ylabel('Freq')

plt.tight_layout()
plt.show()
print("Yes")
```
