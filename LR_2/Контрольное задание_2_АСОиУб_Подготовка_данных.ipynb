## Повторяющиеся значения / Замена значений / Применение функций

### Повторяющиеся значения

```python
# добавление дублирующих строк
df = pd.concat([df]*2, ignore_index=True)
# опрделение дублирующих строк
df[df.duplicated()]
# удаляем дублирующие строки, каждый раз оставляя первое из дублирующихся наблюдений
df.drop_duplicates(keep='first')
# удаляем дублирующие строки, каждый раз оставляя последнее из дублирующихся наблюдений
df.drop_duplicates(keep='last')
# добавляем столбец v
df['v']=0
# но если мы укажем, что нужно удалить дублирующие строки с учетом значений в двух столбцах
df.drop_duplicates(subset={'carat','color'},keep='first',inplace=True)
```

### Замена значений

сопоставляем значения двух числовых столбцов

если между значениями двух столбцов не будет найдено соотвествие, будет выдано значение NaN


замена одного значения на другое

заменяем все значения на новые

заменяем элементы, используя в качестве аргумента словарь

создаем датафрейм с двумя столбцами

задаем разные заменяемые значения для каждого столбца

```python
# сопоставляем значения двух числовых столбцов если между значениями двух столбцов не будет найдено соотвествие, будет выдано значение NaN
df_merge = pd.merge(df['carat'], df['color'], left_index=True, right_index=True)
df_merge
# замена одного значения на другое
df_merge['color'].replace()
df_merge
```

```python
# максимальные значения
df.describe().max()
# # индекс строки, где максимальные значения
df.select_dtypes(include='number').idxmax()
# Среднее значение для всех столбцов датафрейма
df.describe().mean()
# усредненное значение по всем столбцам для каждой строки
df.describe().mean(axis=1)
# # медиана значений для каждого столбца
df.describe().median()
# мода для столбца мод может быть несколько, поэтому результат операции - Series
df.mode()
# вычисляем дисперсию значений в каждом столбце
df.select_dtypes(include='number').var()
# вычисляем среднеквадратичное отклонение
df.select_dtypes(include='number').std()
# вычисляем ковариацию между MSFT и AAPL
print("ковариация:", df['x'].cov(df['y']))
# вычисляем корелляцию между MSFT и AAPL
print("корелляция:", df['x'].corr(df['y']))
```
